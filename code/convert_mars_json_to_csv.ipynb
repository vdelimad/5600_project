```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# json path
json_file_path = '../../data/mars_weather/435_3476_mars_weather.json'

# open
with open(json_file_path, 'r') as json_file:
    json_data = json.load(json_file)

# convert to df
json_data = pd.DataFrame(json_data)
json_data.fillna(value=np.nan, inplace=True)

# fix data types
json_data['id'] = json_data['id'].astype(int)
json_data['terrestrial_date'] = pd.to_datetime(json_data['terrestrial_date'])


# check
display(json_data)

```

```python
# path
csv_file_path = '../../data/mars_weather/1_1977_mars_weather.csv'

# read
csv_data = pd.read_csv(csv_file_path)

# fixes
csv_data.rename(columns={'month': 'season'}, inplace=True)
csv_data['terrestrial_date'] = pd.to_datetime(csv_data['terrestrial_date'])

# check
display(csv_data)

```

```python
# path
missing_file_path = '../../data/mars_weather/mars_missing.csv'

# bring in
missing_data = pd.read_csv(missing_file_path)

# fixes
missing_data['terrestrial_date'] = pd.to_datetime(missing_data['terrestrial_date'])


# check
display(missing_data)

```

```python
# concats
mars_weather_full = pd.concat([json_data, csv_data], ignore_index=True)
mars_weather_full = pd.concat([missing_data, mars_weather_full], ignore_index=True)

# drop duplicates
mars_weather_full.drop_duplicates(inplace=True)
mars_weather_full.reset_index(drop=True, inplace=True)

# remove before year 32
condition = mars_weather_full['terrestrial_date'] > '2013-07-31'
mars_weather_full = mars_weather_full[condition]

# fix column names
mars_weather_full.rename(columns={'min_temp': 'min_air_temp_c'}, inplace=True)
mars_weather_full.rename(columns={'max_temp': 'max_air_temp_c'}, inplace=True)

# drop unnecesary cols
mars_weather_full = mars_weather_full.drop(['id', 'wind_speed', 'atmo_opacity'], axis=1)


# check
display(mars_weather_full)
```

```python
# plot
plt.plot(mars_weather_full['terrestrial_date'], mars_weather_full['pressure'])

plt.xlabel('terrestrial_date')
plt.ylabel('pressure')

plt.show()
```

```python
# plot
plt.plot(mars_weather_full['terrestrial_date'], mars_weather_full['min_air_temp_c'])
plt.xlabel('terrestrial_date')
plt.ylabel('max_air_temp_c')

plt.show()
```

```python
# save
output_csv_path = '../../data/mars_weather/mars_weather_data_full.csv'
mars_weather_full.to_csv(output_csv_path, index=False)

```
