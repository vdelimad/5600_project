```python
import pandas as pd
```

```python
treasury_data = pd.read_csv(f'../../../data/raw/public_sector/treasury_data/monthly_treasury_statement_nasa_20150331_20240131.csv')



display(treasury_data.shape)
display(treasury_data.head())

```

```python
treasury_data = treasury_data[['Record Date', 'Current Month Receipt or Outlay Amount']]

treasury_data.head()
```

```python
treasury_data.columns
```

```python
treasury_data.rename(columns={'Record Date': 'date', 'Current Month Receipt or Outlay Amount': 'outlay_amount'}, inplace=True)
treasury_data['date'] = pd.to_datetime(treasury_data['date'])

treasury_data.sort_values(by='date', ascending=True, inplace=True)

treasury_data.head()
```

```python

treasury_data.dtypes
```

```python
treasury_data.to_csv(f'../../../data/processed/public_sector/treasury_data/monthly_treasury_statement_nasa.csv', index=False)


```
