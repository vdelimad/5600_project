```python
import pandas as pd
```

```python
# json path
spaceflight_data = pd.read_csv(f'../../../data/raw/global_involvement/space_missions/space_corrected.csv')

spaceflight_data
```

```python
spaceflight_data.columns
```

```python
cols_to_rename = {
    'Company Name': 'company_name',
    'Location': 'location',
    'Datum': 'date',
    'Detail': 'detail',
    'Status Rocket': 'status_rocket',
    'Rocket': 'rocket',
    'Status Mission': 'status_mission',
     
}

spaceflight_data.rename(columns=cols_to_rename, inplace=True)

spaceflight_data
```

```python
# splitting the column
spaceflight_data[['address', 'country']] = spaceflight_data['location'].str.rsplit(',', n=1, expand=True)

spaceflight_data.drop(['location'], axis=1, inplace=True)

# stripping any leading/trailing whitespace
spaceflight_data['address'] = spaceflight_data['address'].str.strip()
spaceflight_data['country'] = spaceflight_data['country'].str.strip()

spaceflight_data
```

```python
# find rows where 'date' does not end with 'UTC'
mask = ~spaceflight_data['date'].str.endswith('UTC')


# Append ' 0:00 UTC' to these rows
spaceflight_data.loc[mask, 'date'] = spaceflight_data.loc[mask, 'date'] + ' 0:00 UTC'

```

```python
spaceflight_data['date'] = pd.to_datetime(spaceflight_data['date'], format='%a %b %d, %Y %H:%M %Z')

spaceflight_data.dtypes
```
