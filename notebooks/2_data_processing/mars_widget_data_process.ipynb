```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
# bring in data

mars_weather_data = pd.read_csv(f'../../../data/raw/public_sector/mars_weather_from_widget/mars_weather_from_widget_raw.csv')

display(mars_weather_data.shape)
display(mars_weather_data.head())
```

```python
# bring in mars years data and covert to dt

mars_years = pd.read_csv(f'../../../data/raw/martian_years.csv')
for column in mars_years.columns[1:]:
    mars_years[column] = pd.to_datetime(mars_years[column], dayfirst=True)

display(mars_years)
```

```python
# rename
mars_weather_data = mars_weather_data.rename(columns={'sol': 'curiosity_sol'})


# replace all the Value not available with nan
mars_weather_data['relative_humidity_current'] = mars_weather_data['relative_humidity_current'].replace('Value not available', np.nan)
mars_weather_data['wind_current'] = mars_weather_data['wind_current'].replace('Value not available', np.nan)

# check
display(mars_weather_data.head())
```

```python
# dropping columns where all values are NaN
mars_weather_data = mars_weather_data.dropna(axis=1, how='all')

# check
display(mars_weather_data.head())
```

```python
# drop unneded cols
mars_weather_data.drop(columns=['wind_unit', 'relative_humidity_unit', 'atmospheric_opacity_current', 'air_temperature_unit', 'ground_temperature_unit', 'pressure_unit'], inplace=True)

# fix uv col
mars_weather_data = mars_weather_data.replace('Value not available', np.nan)
mars_weather_data['ultraviolet_radiation_current'] = mars_weather_data['ultraviolet_radiation_current'].replace('Not available', np.nan)


# rename
mars_weather_data = mars_weather_data.rename(columns={'mars_season': 'mars_month', 'air_temperature_max': 'air_temperature_max_celsius', 'air_temperature_min': 'air_temperature_min_celsius', 'ground_temperature_max': 'ground_temperature_max_celsius', 'ground_temperature_min': 'ground_temperature_min_celsius', 'pressure_current': 'pressure_current_pascals'})


# remove text from integer cols
mars_weather_data['curiosity_sol'] = mars_weather_data['curiosity_sol'].str.extract('(\d+)').astype(int)
mars_weather_data['mars_month'] = mars_weather_data['mars_month'].str.extract('(\d+)').astype(int)

# check
display(mars_weather_data.head())
```

```python
# convert col to date
mars_weather_data['earth_date_utc'] = pd.to_datetime(mars_weather_data['earth_date_utc'])

# convert cols to int
columns_to_convert_to_int = ['mars_month', 'solar_longitude_degrees', 'air_temperature_max_celsius', 'air_temperature_min_celsius', 'ground_temperature_max_celsius', 'ground_temperature_min_celsius', 'pressure_current_pascals']
mars_weather_data[columns_to_convert_to_int] = mars_weather_data[columns_to_convert_to_int].astype('Int64')

# convert times
mars_weather_data['sunrise_and_sunset_max'] = pd.to_datetime(mars_weather_data['sunrise_and_sunset_max'], format='%H:%M').dt.time
mars_weather_data['sunrise_and_sunset_min'] = pd.to_datetime(mars_weather_data['sunrise_and_sunset_min'], format='%H:%M').dt.time

# lowercase
mars_weather_data['ultraviolet_radiation_current'] = mars_weather_data['ultraviolet_radiation_current'].str.lower()


# check
display(mars_weather_data.head())
```

```python

def find_mars_year_and_season(earth_date, solar_longitude_degrees):
    
    for i, row in mars_years.iterrows():
        spring_equinox = row['spring_equinox']
        if i < len(mars_years) - 1:
            next_spring_equinox = mars_years.iloc[i + 1]['spring_equinox']
        else:
            next_spring_equinox = spring_equinox + pd.DateOffset(years=2)  # Approximation
        
        # check if date falls within this mars year range
        if spring_equinox <= earth_date < next_spring_equinox:
            mars_year = row['mars_year']
            break
    else:
        # handle errors
        return None, None

    # determine season based on longitude
    if 0 <= solar_longitude_degrees < 90:
        season = "spring"
    elif 90 <= solar_longitude_degrees < 180:
        season = "summer"
    elif 180 <= solar_longitude_degrees < 270:
        season = "autumn"
    else:
        season = "winter"
    
    return mars_year, season

# apply
mars_weather_data['mars_year'], mars_weather_data['season'] = zip(*mars_weather_data.apply(lambda row: find_mars_year_and_season(row['earth_date_utc'], row['solar_longitude_degrees']), axis=1))
```

```python
# individual data errors fix
mars_weather_data.loc[mars_weather_data['curiosity_sol'] == 3693, 'mars_year'] = 36
```

```python

# fix col order
first_cols = ['curiosity_sol', 'earth_date_utc', 'mars_year', 'mars_month', 'solar_longitude_degrees', 'season']
remaining_cols_sorted = sorted(mars_weather_data.columns.drop(first_cols))
cols_sorted = first_cols + remaining_cols_sorted

mars_weather_data = mars_weather_data[cols_sorted]

# check
mars_weather_data.head()

```
