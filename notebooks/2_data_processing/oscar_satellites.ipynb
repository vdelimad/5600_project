```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
oscar_data = pd.read_csv(f'../../../data/raw/space_infrastructure/satellite_data/oscar_satellites_2024_02_05.csv')

oscar_data
```

```python
oscar_data.columns
```

```python
cols_to_rename = {
    
    'Acronym': 'acronym',
    'SAT ID': 'sat_id',
    'Launch': 'launch_date',
    '(expected) EOL': 'expected_eol_date',
    'Satellite Programme': 'satellite_programme',
    'Agencies': 'agencies',
    'Orbit': 'orbit',
    'Altitude': 'altitude',
    'Longitude': 'longitude',
    'Inclination': 'inclination',
    'Ect': 'ect',
    'Sat status': 'sat_status',
    'Payload': 'payload',
    'Last update': 'last_update'
    
    }
      

oscar_data.rename(columns=cols_to_rename, inplace=True)

oscar_data


```

```python
oscar_data['payload'] = oscar_data['payload'].apply(lambda x: x.split('\n') if isinstance(x, str) else [])

oscar_data['agencies'] = oscar_data['agencies'].apply(lambda x: x.strip('\n').split('\n') if isinstance(x, str) else [])

oscar_data
```

```python
oscar_data = oscar_data[oscar_data['launch_date'] != 'TBD']

oscar_data
```

```python
oscar_data.loc[:, 'launch_date'] = pd.to_datetime(oscar_data['launch_date'], format='%d/%m/%Y')
oscar_data.loc[:, 'expected_eol_date'] = pd.to_datetime(oscar_data['expected_eol_date'], format='%d/%m/%Y')
oscar_data.loc[:, 'last_update'] = pd.to_datetime(oscar_data['last_update'], format='%d/%m/%Y %H:%M')

display(oscar_data.dtypes)
display(oscar_data.head())


```

```python
oscar_data.to_csv(f'../../../data/processed/space_infrastructure/satellite_data/oscar_satellites.csv', index=False)


```
